### Applying PCA and plotting 


Now we will
1. Run PCA to visualize and improve results for 5 clusters
2. Identify the best number ofcomponents


```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

X = X_train
y_num = predictions

target_names = ["Cluster 0","Cluster 1","Cluster 2", "Cluster 3","Cluster 4"]

pca = PCA(n_components=2, random_state = 453)
X_r = pca.fit(X).transform(X)


# PRECENTAGE OF VARIANCE EXPLAINED FOR EACH COMPONENTS
print('Explained variance ratio (first two components): %s' % str(pca.explained_variance_ratio_))

# PLOTTING DATA POINTS
plt.figure()
plt.figure(figsize=(12,8))
colors = ['navy', 'turquoise', 'darkorange', 'red', 'purple']
lw = 0.1


for color, i, target_name in zip(colors, [0, 1, 2, 3,4], target_names):
    plt.scatter(X_r[y_num == i, 0], X_r[y_num == i, 1], color=color, alpha=.8, lw=lw,label=target_name)
    
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.6)   
plt.title('PCA of 2 Items')
plt.show()
```

```python
# IDENTIFYING THE "BEST" NUMBER OF COMPONENTS : TRYING WITH DIMENTIONALITY REDUCTION & K-MEANS
n_components = X.shape[1]

# RUNNING PCA WITH ALL COMPONENTS
pca = PCA(n_components=n_components, random_state = 453)
X_r = pca.fit(X).transform(X)

# CALCULATING THE 95% VARIANCE
total_variance = sum(pca.explained_variance_)
print("Total Variance in our dataset is: ", total_variance)
var_95 = total_variance * 0.95
print("The 95% variance we want to have is: ", var_95)
print("")

# CREATING A DF WITH THE COMPONENTS & EXPLAINED VARIANCE
a = zip(range(0,n_components), pca.explained_variance_)
a = pd.DataFrame(a, columns=["PCA Comp", "Explained Variance"])

# TRYING TO HIT 95% VARIANCE
print("Variance explain with 4 n_compononets: ", sum(a["Explained Variance"][0:4]))
print("Variance explain with 8 n_compononets: ", sum(a["Explained Variance"][0:8]))
print("Variance explain with 9 n_compononets: ", sum(a["Explained Variance"][0:9]))
print("Variance explain with 10 n_compononets: ", sum(a["Explained Variance"][0:10]))
print("Variance explain with 15 n_compononets: ", sum(a["Explained Variance"][0:15]))
print("Variance explain with 18 n_compononets: ", sum(a["Explained Variance"][0:18]))
print("Variance explain with 21 n_compononets: ", sum(a["Explained Variance"][0:21]))

# PLOTTING DATA POINTS
plt.figure(1, figsize=(14, 10))
plt.plot(pca.explained_variance_ratio_, linewidth=2, c="r")
plt.xlabel('no_of_components')
plt.ylabel('explained_ratio_')

# PLOTTING LINE WITH 95% E.V.
plt.axvline(9,linestyle=':', label='n_components - 95% explained', c ="green")
plt.legend(prop=dict(size=12))

# ADDING ARROW POINTER
plt.annotate(' 9 eigenvectors used to explain 95% variance', xy=(9, pca.explained_variance_ratio_[9]), 
             xytext=(9, pca.explained_variance_ratio_[9]),
            arrowprops=dict(facecolor='green', shrink=0.05))

plt.show()
```

Note: from the above we can see that 9 eigen vectors can be used to represent 95% variance

## Running Kmeans clustering on the entire dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Creating a random dataset
np.random.seed(42)
X_train = np.random.rand(100, 2)

# Creating KMeans object with 5 clusters
kmeans = KMeans(n_clusters=5, random_state=2)

# Fitting the model to the data
kmeans = kmeans.fit(X_train)

# Predicting the cluster labels for each data point
predictions = kmeans.predict(X_train)

# Calculating cluster count
unique, counts = np.unique(predictions, return_counts=True)
counts = counts.reshape(1,5)

# Creating Pandas dataframe for cluster counts
countscldf = pd.DataFrame(counts, columns=["Cluster 0", "Cluster 1", "Cluster 2", "Cluster 3", "Cluster 4"])

# Plotting the clusters
fig, ax = plt.subplots(figsize=(8,6))
plt.scatter(X_train[:,0], X_train[:,1], c=predictions, cmap='cool')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='*')
plt.title('K-means Clustering Plot')
plt.xlabel('X')
plt.ylabel('Y')

# Adding cluster counts as a table
ax.table(cellText=countscldf.values,
         rowLabels=countscldf.index,
         colLabels=countscldf.columns,
         loc='bottom')
plt.show()
```

```python
labels = kmeans.labels_
n_clusters = len(set(labels))
print('Number of clusters:', n_clusters)
```

```python
from sklearn.metrics import silhouette_score

# Calculate the silhouette score
silhouette_avg = silhouette_score(X_train,labels)   
print('Silhouette Score of Kmeans:', silhouette_avg)
```

## Applying Agglomerative clustering

```python
from scipy.cluster.hierarchy import linkage, fcluster
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("C:\\Users\\anura\\Desktop\\SYMBIOSIS\\DATASETS\\ECommerce_consumer behaviour.csv")

# Drop any missing values
data = data.dropna()
data.drop(['department','product_name'],axis=1,inplace=True)

# Convert the dataframe to a numpy array
X_full = data.values

# Set the number of data points to subsample
n_samples = 20000

# Randomly subsample the data
rng = np.random.default_rng(seed=42)
idx = rng.choice(X_full.shape[0], size=n_samples, replace=False)
X = X_full[idx, :]

# Compute the linkage matrix using the ward method
Z = linkage(X, method='ward')

# Set the number of clusters
k = 3

# Assign data points to clusters based on the linkage matrix
y_pred = fcluster(Z, k, criterion='maxclust')

# Add the predicted cluster labels to the dataframe
data_subsample = data.iloc[idx, :].copy()
data_subsample['Cluster'] = y_pred

# Print the number of samples in each cluster
print(data_subsample['Cluster'].value_counts())

# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title('Agglomerative Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
score = silhouette_score(X, y_pred)
print(f"Silhouette Score of agglomerative: {score}")
```
